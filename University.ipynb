```python
import pandas as pd

file_path = "/content/2022년 학교별 고등교육기관 취업통계 - 학교별.csv"

df = pd.read_csv(file_path, header=[3, 4])


new_columns = []

for col_level0, col_level1 in df.columns:
    col0_str = str(col_level0)
    col1_str = str(col_level1)

    if 'Unnamed' in col0_str and 'Unnamed' in col1_str:
        new_columns.append(f"{col0_str}_{col1_str}")

    elif 'Unnamed' in col0_str:
        new_columns.append(col1_str)

    elif 'Unnamed' in col1_str or pd.isna(col_level1) or col_level1 == '':
        new_columns.append(col0_str)


    else:
        new_columns.append(f"{col0_str}_{col1_str}")


df.columns = new_columns

columns_to_drop = [col for col in df.columns if 'Unnamed' in str(col)]
df = df.drop(columns=columns_to_drop, errors='ignore')

print(df.head())
```
