```python
import pandas as pd

df_industry = pd.read_csv('후쿠오카_산업별_취업자수.csv')
df_combined = pd.read_csv('데이터통합_미완성.csv')

df_merged = pd.merge(
    df_combined,
    df_industry,
    on='연도_반기',
    how='left'  
)

df_merged.to_csv('데이터통합_완성.csv', index=False, encoding='utf-8-sig')

print("병합이 완료되었습니다. '데이터통합_완성.csv' 파일을 확인하세요.")
```

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('데이터통합_완성.csv')

# _x 접미사 열 (빈 열) 제거
columns_to_drop = [col for col in df.columns if col.endswith('_x')]
df = df.drop(columns=columns_to_drop)

# _y 접미사 열 이름을 원래 이름으로 변경
columns_to_rename = {
    '건설업 (F)_y': '건설업 (F)',
    '광·제조업 (B,C)_y': '광·제조업 (B,C)',
    '농업, 임업 및 어업 (A)_y': '농업, 임업 및 어업 (A)',
    '도소매·숙박음식업 (G,I)_y': '도소매·숙박음식업 (G,I)',
    '사업·개인·공공서비스 및 기타 (E,L~U)_y': '사업·개인·공공서비스 및 기타 (E,L~U)',
    '전기·운수·통신·금융 (D,H,J,K)_y': '전기·운수·통신·금융 (D,H,J,K)',
    '정보통신 (J)_y': '정보통신 (J)'
}
df = df.rename(columns=columns_to_rename)

# Unnamed: 0 열 제거 (필요한 경우)
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# 정리된 데이터 저장
df.to_csv('데이터통합_최종.csv', index=False, encoding='utf-8-sig')

print("데이터 정리가 완료되었습니다. '데이터통합_최종.csv' 파일을 확인하세요.")
```

```python
import pandas as pd

# CSV 파일 읽기
df_final = pd.read_csv('데이터통합_최종.csv', encoding='utf-8-sig')
df_half_year = pd.read_csv('후쿠오카_고용률_반기별.csv', encoding='utf-8-sig')

# 데이터통합_최종.csv에서 '15세 이상 인구 (명)', '취업자 수 (명)', '고용률 (%)' 열 제거
columns_to_drop = ['고용률(%)']
df_final = df_final.drop(columns=[col for col in columns_to_drop if col in df_final.columns])

# 후쿠오카_고용률_반기별.csv에서 필요한 열만 선택
df_half_year = df_half_year[['연도_반기','고용률(%)']]

# 연도_반기 열을 기준으로 병합
df_merged = pd.merge(df_final, df_half_year, on='연도_반기', how='left')

# 결과 CSV 파일로 저장
df_merged.to_csv('데이터통합_최종_병합.csv', index=False, encoding='utf-8-sig')

# 결과 확인
print(df_merged)
```
