```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (Windows: Malgun Gothic, macOS/Linux: AppleGothic 또는 NanumGothic)
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS
# plt.rcParams['font.family'] = 'NanumGothic'  # Linux
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# CSV 파일 읽기
df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

# 1. 순이동수와 취업자 총계 추이 (선 그래프)
fig, ax1 = plt.subplots(figsize=(12, 6))

# 순이동수 그래프 (왼쪽 Y축)
ax1.plot(df['연도_반기'], df['순이동수'], color='b', marker='o', label='순이동수')
ax1.set_xlabel('연도_반기')
ax1.set_ylabel('순이동수', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_xticklabels(df['연도_반기'], rotation=45)

# 취업자 총계 그래프 (오른쪽 Y축)
ax2 = ax1.twinx()
ax2.plot(df['연도_반기'], df['취업자_총계'], color='r', marker='s', label='취업자 총계')
ax2.set_ylabel('취업자 총계', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# 제목과 범례
plt.title('연도별 순이동수와 취업자 총계 추이')
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
plt.tight_layout()
plt.savefig('순이동수_취업자총계_추이.png')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

fig, ax1 = plt.subplots(figsize=(12, 6))
sns.lineplot(data=df, x='연도_반기', y='순이동수', ax=ax1, color='b', marker='o', label='순이동수')
ax1.set_ylabel('순이동수', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
sns.lineplot(data=df, x='연도_반기', y='고용률 (%)', ax=ax2, color='r', marker='s', label='고용률 (%)')
ax2.set_ylabel('고용률 (%)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('연도별 순이동수와 고용률 추이')
ax1.set_xticklabels(df['연도_반기'], rotation=45)
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
plt.tight_layout()
plt.savefig('순이동수_고용률_추이.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

industries = ['건설업 (F)', '광·제조업 (B,C)', '도소매·숙박음식업 (G,I)', '사업·개인·공공서비스 및 기타 (E,L~U)']
df_plot = df[['연도_반기'] + industries].set_index('연도_반기')

ax = df_plot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title('연도별 산업별 취업자 수 (스택형)')
plt.xlabel('연도_반기')
plt.ylabel('취업자 수')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('산업별_취업자_스택형.png')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

fig, ax1 = plt.subplots(figsize=(12, 6))
sns.lineplot(data=df, x='연도_반기', y='정보통신 (J)', ax=ax1, color='b', marker='o', label='정보통신 (J)')
ax1.set_ylabel('정보통신 취업자 수', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
sns.lineplot(data=df, x='연도_반기', y='취업자_총계', ax=ax2, color='r', marker='s', label='취업자 총계')
ax2.set_ylabel('취업자 총계', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('연도별 정보통신 산업과 취업자 총계 추이')
ax1.set_xticklabels(df['연도_반기'], rotation=45)
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
plt.tight_layout()
plt.savefig('정보통신_취업자총계_추이.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

plt.figure(figsize=(12, 6))
plt.bar(df['연도_반기'], df['전입수'], label='전입수', color='skyblue', width=0.4)
plt.bar(df['연도_반기'], df['전출수'], label='전출수', color='salmon', width=0.4, alpha=0.5, bottom=df['전입수'])
plt.title('연도별 전입수와 전출수 비교')
plt.xlabel('연도_반기')
plt.ylabel('인구 수')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('전입수_전출수_비교.png')
plt.show()
```
