```python
import pandas as pd
import glob #디렉터리 내의 특정 확장자나 이름을 가진 파일들을 한꺼번에 찾는 기능을 제공
import os

folder_path = '/content/Untitled Folder'

txt_files = glob.glob(os.path.join(folder_path, '*.txt')) #.txt 파일 가져오기

df1 = [] #데이터프레임 저장할 리스트

for file in txt_files:
  df = pd.read_csv(file, encoding='utf-8')
  df1.append(df) #읽은 데이터프레임을 리스트에 추가

combined_df = pd.concat(df1, ignore_index=True) #행 방향으로 데이터프레임 하나로 합치기

#연도 열 기준으로 오름차순 정리, 인덱스를 초기화
combined_df = combined_df.sort_values(by=['연도', '월'], ascending=True).reset_index(drop=True)

# 반기 구분 (1~6월: 1/2, 7~12월: 2/2)
combined_df['반기'] = combined_df['월'].apply(lambda x: '1/2' if x <= 6 else '2/2')
combined_df['연도반기'] = combined_df['연도'].astype(str) + '.' + combined_df['반기']

# 연도반기별로 취업_건수 합계 계산
grouped_df = combined_df.groupby('연도반기', as_index=False)['취업_건수'].sum()

# 결과 출력
print(grouped_df)
```
