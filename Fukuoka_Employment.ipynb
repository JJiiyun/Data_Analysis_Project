```python
import pandas as pd
import glob
import os

folder_path = '/content/Untitled Folder'

all_txt_files = glob.glob(os.path.join(folder_path, '*.txt'))

df_list = []

for file in all_txt_files:
    try:
        df = pd.read_csv(file, encoding='utf-8')
    except Exception as e:
        print(f"Error reading file {file}: {e}")
        continue

    df['파일명'] = os.path.basename(file)
    df_list.append(df)


if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    print("Combined DataFrame Head:")
    print(combined_df.head())
else:
    print("No text files found or read successfully in the specified folder.")

if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)

    combined_df = combined_df.sort_values(by='연도', ascending=True).reset_index(drop=True)

    print("Combined DataFrame Head (Sorted by 연도):")
    print(combined_df.head())
else:
    print("No text files found or read successfully in the specified folder.")
```
