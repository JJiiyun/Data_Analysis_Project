```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 폰트 설정 (Windows: Malgun Gothic, macOS/Linux: AppleGothic 또는 NanumGothic)
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS
# plt.rcParams['font.family'] = 'NanumGothic'  # Linux
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# CSV 파일 읽기
df = pd.read_csv('데이터통합_최종_병합.csv', encoding='utf-8')

# 분석할 변수 선택
selected_cols = ['고용률 (%)', '순이동수', '취업자_총계', '전입수', '전출수', '정책시행']
numeric_df = df[selected_cols]

# 피어슨 상관계수 행렬 계산
corr_matrix = numeric_df.corr(method='pearson')

# 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, 
            annot=True,  # 상관계수 값 표시
            cmap='coolwarm',  # 색상 스킴 (빨강: 양의 상관, 파랑: 음의 상관)
            vmin=-1, vmax=1,  # 색상 범위 고정
            center=0,  # 0을 중앙으로
            fmt='.2f',  # 소수점 2자리
            square=True,  # 셀을 정사각형으로
            cbar_kws={'label': '피어슨 상관계수'})
plt.title('선택 변수 간 피어슨 상관계수 히트맵')
plt.tight_layout()
plt.savefig('선택변수_상관계수_히트맵.png')
plt.show()
```
