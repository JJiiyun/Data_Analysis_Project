```python
#다중헤더 header=[0,1]
df = pd.read_csv('/content/시군구별_이동자수_20250607160508.csv', encoding='cp949', header=[0,1])
df.head()
```

```python
df.set_index(('행정구역(시군구)별', '행정구역(시군구)별'), inplace=True)
df.head()
```

```python
seoul = df.loc['서울특별시'] #서울특별시 데이터만 선택

total_in_cols = [col for col in df.columns if col[1] == '총전입 (명)'] #총전입인 컬럼들만 선택
total_out_cols = [col for col in df.columns if col[1] == '총전출 (명)']


seoul_in = seoul[total_in_cols]
seoul_out = seoul[total_out_cols]


dates_str = [col[0] for col in total_in_cols] #('2015.01', '총전입 (명)')
dates = pd.to_datetime(dates_str, format='%Y.%m')

seoul_in.index = dates
seoul_out.index = dates

plt.figure(figsize=(12,6))
plt.plot(seoul_in.values, label='총전입 (명)', marker='o')
plt.plot(seoul_out.values, label='총전출 (명)', marker='x')
plt.xticks(ticks=np.arange(len(dates)), labels=dates.strftime('%Y-%m'), rotation=45)
plt.title('서울특별시 월별 총전입·총전출 (2015~2019)')
plt.xlabel('날짜')
plt.ylabel('명')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
